```python
from datasets import load_dataset

```

```python
trec = load_dataset("trec")
clinic = load_dataset("clinc_oos", 'plus')
banking = load_dataset("banking77")
nlu = load_dataset("nlu_evaluation_data")
nlu
```

```python
from matplotlib import pyplot as plt
import numpy as np

def graph(data, savename):
    plt.hist(list(data), len(list(set(data))), alpha=0.5, edgecolor = 'black')
    plt.axis('off')
    plt.savefig(f"sparklines/{savename}.pdf", bbox_inches="tight")
    plt.show()
    
    
data = list(trec['train']['coarse_label'])
graph(data, "trec")

data = list(trec['train']['fine_label'])
graph(data, "trecfine")

data = list(clinic['train']['intent'])
graph(data, "clinic")


data = list(banking['train']['label'])
graph(data, "banking")

data = list(nlu['train']['label'])


graph(data, "nlu")

```
